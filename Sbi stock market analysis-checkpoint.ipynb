### Price difference of each day

```python
sbi['Price Diff'] = (sbi['Close'] - (sbi['Close'].shift(1)).fillna(0)).shift(-1)
sbi.head()
```

```python
sbi1['Price Diff'] = (sbi1['Close'] - (sbi1['Close'].shift(1)).fillna(0)).shift(-1)
sbi1.head()
```

### New columns SMA(20), SMA(50), EMA(20) & EMA(50)

```python
sbi['SMA_20'] = sbi['Close'].rolling(20).mean()

sbi['SMA_50'] = sbi['Close'].rolling(50).mean()

sbi.head()
```

```python
sbi['EMA_20'] = sbi['Close'].ewm(20).mean()

sbi['EMA_50'] = sbi['Close'].ewm(50).mean()
sbi.head()
```

```python
sbi1['SMA_20'] = sbi1['Close'].rolling(20).mean()

sbi1['SMA_50'] = sbi1['Close'].rolling(50).mean()

sbi1.head()
```

```python
sbi1['EMA_20'] = sbi1['Close'].ewm(20).mean()

sbi1['EMA_50'] = sbi1['Close'].ewm(50).mean()
sbi1.head()
```

### SMA(20) and SMA(50) with close price for one year data graph

### Overall profit and the graph of profit 

#### when close price is greater than EMA(20) and sell when it is less than SMA(20)

```python
n = sbi.shape[0]
price = (sbi.dropna()).reset_index()
del price['index']
print(n)
sbi.head()
```

```python
for i in range(n):
    if sbi.loc[i,'Close'] > sbi.loc[i,'EMA_20']:
        sbi.loc[i,'Buy'] = 1
    if sbi.loc[i,'Close'] < sbi.loc[i,'SMA_20']:
        sbi.loc[i,'Buy'] = 0
sbi.head()
```

```python
sbi['Profit'] = sbi['Price Diff']*sbi['Buy']
sbi.head()
```

```python
Profit = sbi['Profit'].sum()
print('Profit =',Profit)
```

#### when close price is greater than EMA(50) and sell when it is less than SMA(50)

```python
for i in range(n):
    if sbi.loc[i,'Close'] > sbi.loc[i,'EMA_50']:
        sbi.loc[i,'Buy'] = 1
    if sbi.loc[i,'Close'] < sbi.loc[i,'SMA_50']:
        sbi.loc[i,'Buy'] = 0
sbi.head()
```

```python
sbi['Profit'] = sbi['Price Diff']*sbi['Buy']
Profit = sbi['Profit'].sum()
print('Profit =',Profit)
```
